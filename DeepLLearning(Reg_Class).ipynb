```python
#imports
from sklearn.datasets import make_circles
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd

tf.random.set_seed(42)
```

```python
X,y = make_circles(100, noise=0.03, random_state=42)
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
X = tf.convert_to_tensor(X)
y = tf.convert_to_tensor(y)
```

```python
import numpy as np

def plot_decision_boundary(model, X, y):
  X=X.numpy()
  y=y.numpy()
  """
  Plots the decision boundary created by a model predicting on X.
  This function has been adapted from two phenomenal resources:
   1. CS231n - https://cs231n.github.io/neural-networks-case-study/
   2. Made with ML basics - https://github.com/GokuMohandas/MadeWithML/blob/main/notebooks/08_Neural_Networks.ipynb
  """
  # Define the axis boundaries of the plot and create a meshgrid
  x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
  y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
  xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                       np.linspace(y_min, y_max, 100))

  # Create X values (we're going to predict on all of these)
  x_in = np.c_[xx.ravel(), yy.ravel()] # stack 2D arrays together: https://numpy.org/devdocs/reference/generated/numpy.c_.html

  # Make predictions using the trained model
  y_pred = model.predict(x_in)

  # Check for multi-class
  if model.output_shape[-1] > 1: # checks the final dimension of the model's output shape, if this is > (greater than) 1, it's multi-class
    print("doing multiclass classification...")
    # We have to reshape our predictions to get them ready for plotting
    y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
  else:
    print("doing binary classifcation...")
    y_pred = np.round(np.max(y_pred, axis=1)).reshape(xx.shape)

  # Plot decision boundary
  plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation=tf.keras.activations.relu, input_shape=(2,)),
    tf.keras.layers.Dense(10, activation = 'relu'),
    tf.keras.layers.Dense(1, activation=tf.keras.activations.sigmoid)
])

model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])

lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))

history = model.fit(tf.expand_dims(X[:80],axis=-1),y[:80], epochs=100, verbose=0, callbacks=[lr_scheduler])
```

```python
pd.DataFrame(history.history).plot()
```

```python
lrs = 1e-4*(10**(np.arange(100)/20))
plt.semilogx(lrs, history.history['loss'])
```

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import fashion_mnist

tf.random.set_seed(42)
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
model1 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(10, activation="softmax")
])

model1.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.008),
              metrics=['accuracy'])

lr_seq = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-3 * 10**(epoch/20))

model1.fit(X_train, y_train, epochs=20)
```

```python
lrs = 1e-3*10**(np.arange(40)/20)
plt.semilogx(lrs, model1.history.history['loss'])
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, model1.predict(X_test).argmax(axis=1))
print(cm)
```

```python
import random
import time
def get_random_result(model, images, labels, classes):
  i = random.randint(0, len(images))
  image = images[i]
  pred_label = model.predict(image.reshape(1,28,28)).argmax()
  true_label = labels[i]
  if true_label==pred_label:
    color = 'green'
  else:
    color = 'red'
  plt.imshow(image)
  plt.title(f'Predicted: {classes[pred_label]}, True: {classes[true_label]}', color=color)
```

```python
get_random_result(model1, X_train, y_train, class_names)
```
